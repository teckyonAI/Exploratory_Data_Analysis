 ## **3: Exploarity Data Analysis** <a class="anchor" id="3"></a>


### **Univarient Analysis** <a class="anchor" id="3.1"></a>

#### **Categorical Feature** <a class="anchor" id="3.1.1"></a>

```python
categorical_feature = [feature for feature in data.columns if data[feature].dtypes =="O"]
print(len(categorical_feature))
categorical_feature
```

```python
data[categorical_feature].info()
```

##### **Count_Plot** <a class="anchor" id="3.1.1.1"></a>

```python
plt.figure(figsize=(20, 10))

for (i, x) in enumerate(categorical_feature):
    
    plt.subplot(2, 2, i + 1)
    sns.countplot(x = data[x] )
    plt.title(f'{x} spread')

plt.tight_layout()
```

##### **Pie_ plot** <a class="anchor" id="3.1.1.2"></a>

```python
plt.figure(figsize=(20, 10))

for (i, x) in enumerate(categorical_feature):
    
    plt.subplot(2, 2, i + 1)
    data[x].value_counts().plot(kind='pie', autopct='%.2f')
    plt.title(f'{x} spread')

plt.tight_layout()
```

#### **Numerical Feature** <a class="anchor" id="3.1.2"></a>

```python
numerical_feature = [feature for feature in data.columns if data[feature].dtypes != 'O']
print("The length of Numerical_values is :",len(numerical_feature))
numerical_feature
```

##### **Histogram** <a class="anchor" id="3.1.2.1"></a>

```python
plt.figure(figsize=(20, 10))

for (i, x) in enumerate(numerical_feature):
    
    plt.subplot(3, 4, i+1)
    sns.histplot(x = data[x],bins=10)
    plt.title(f'{x} spread')
    
plt.tight_layout()
```

##### **Distplot** <a class="anchor" id="3.1.2.2"></a>

```python
plt.figure(figsize=(20, 10))

for (i, x) in enumerate(numerical_feature):
    
    plt.subplot(3, 4, i+1)
    sns.histplot(data[x], color="red", label="100% Equities", kde=True, stat="density", linewidth=0,bins=10)
    plt.title(f'{x} spread')
    
plt.tight_layout()
```

##### **Boxplot** <a class="anchor" id="3.1.2.3"></a>

```python
plt.figure(figsize=(20, 10))

for (i, x) in enumerate(numerical_feature):
    
    plt.subplot(3, 4, i+1)
    sns.boxplot(data = data[x],color='green')
    plt.title(f'{x} spread')
    
plt.tight_layout()
```

### **Bi_varient / Multi_varient** <a class="anchor" id="3.2"></a>

```python
data.head()
```

```python
features =  data.dropna()
```

#### **Scatter_Plot (Numerical - Numerical)** <a class="anchor" id="3.2.1"></a>

```python
plt.figure(figsize=(20, 10))
 
features = data.drop(['case','site','age','Pop','sex'],axis=1)

for (i, x) in enumerate(features):
    
    plt.subplot(3,3,i+1)
    sns.scatterplot(x=data['age'],y=data[x],hue=data['sex'],style = data['Pop'])
    plt.title(f'{x} spread')
    
plt.tight_layout()
```

#### **Bar_Plot (Numerical - Categorical)** <a class="anchor" id="3.2.2"></a>

```python
plt.figure(figsize=(20, 10))
 
features = data.drop(['case','site','Pop','sex'],axis=1)

for (i, x) in enumerate(features):
    
    plt.subplot(3,4,i+1)
    sns.barplot(x=data['Pop'],y=data[x],hue=data['sex'])
    plt.title(f'{x} spread')
    
plt.tight_layout()
```

#### **Box_Plot (Numerical - Categorical)** <a class="anchor" id="3.2.3"></a>

```python
plt.figure(figsize=(20, 10))
 
features = data.drop(['case','site','Pop','sex'],axis=1)

for (i, x) in enumerate(features):
    
    plt.subplot(3,4,i+1)
    sns.boxplot(x=data['Pop'],y=data[x],hue=data['sex'],color='blue')
    plt.title(f'{x} spread')
    
plt.tight_layout()
```

#### **Dist_Plot (Numerical - Categorical)** <a class="anchor" id="3.2.4"></a>

```python

plt.figure(figsize=(20, 10))
 
features = data.drop(['case','site','Pop','sex'],axis=1)

for (i, x) in enumerate(features):
    
    plt.subplot(3,4,i+1)
    sns.distplot(data[data['Pop']=='Vic'][x],hist=False)
    sns.distplot(data[data['Pop']=='other'][x], hist=False)
    plt.title(f'{x} spread')
    
plt.tight_layout()
```

#### **Heat_Map (Categorical - Categorical)** <a class="anchor" id="3.2.5"></a>

```python
sns.heatmap(pd.crosstab(data['Pop'],data['sex']))
```

```python
sns.heatmap(pd.crosstab(data['Pop'],data['age']))
```

```python
# Average age of Pop
(data.groupby('Pop').mean()['age'].plot(kind='bar',color='purple'))
```

#### **Cluster_Map (Categorical - Categorical)** <a class="anchor" id="3.2.6"></a>

```python
sns.clustermap(pd.crosstab(data['Pop'],data["age"]))
```

#### **Pair_Plot** <a class="anchor" id="3.2.7"></a>

```python
sns.pairplot(data,hue='sex')
```

#### **Line_Plot (Numerical - Numerical)** <a class="anchor" id="3.2.8"></a>

```python
plt.figure(figsize=(20, 10))
 
features = data.drop(['age','case','site','Pop','sex'],axis=1)

for (i, x) in enumerate(features):
    
    plt.subplot(3,4,i+1)
    sns.lineplot(data['age'], data[x],hue=data['Pop'],style=data['sex'],color='purple')
    plt.title(f'{x} spread')
    
plt.tight_layout()
```
